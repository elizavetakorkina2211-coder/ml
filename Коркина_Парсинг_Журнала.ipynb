Вернем список ссылок на новости, полученные с текущей страницы

```python
def getPageLinks(page_number):
    # составляем ссылку на страницу поиска
    page = 'https://journal.tinkoff.ru/flows/economics/page/{}'.format(page_number)
    # запрашиваем данные по ней
    response = get(page)
    if not response.ok:
        # если сервер нам отказал, вернем пустой лист для текущей страницы
        return []
    # получаем содержимое страницы
    html = response.content
    soup = BeautifulSoup(html,'html.parser')
    # ищем ссылки на новости
    links = soup.find_all('a', {'class':'link--PVaAl'})
    links = ['https://journal.tinkoff.ru' + link.get('href') for link in links]

    return links
```

```python
#функция, которая будет возращать название статьи, для кого статья,дату,кол-во просмотров,кол-во кокомментариев,автора и лайки
def getProperties(soup):
    article = soup.find('h1',{'class':'articleTitle--pUvzk'}).text.replace('\xa0', ' ').strip()#название статьи

    t = soup.find('div',{'class':'header--W4Uh8'})
    t1 = t.find('a',{'class':'flow--CBXD2'}).text.replace('\xa0', ' ').strip() #для кого новость

    data = soup.find('div',{'class':'dateWrapper--JWrXQ'})
    Data = data.find('time').text.strip() #дата

    views = soup.find('span',{'class':'counter--6lus3'}).text.strip() #просмотры

    comment = soup.find('span',{'class':'content--FZH2E'}).text.strip() #комментарии
    avtor = soup.find('p',{'class':'paragraph--v7OR4'}).text.strip() #автор
    like = soup.find('span',{'class':'counter--IZ6Up'}).text.strip() #лайки

    return article, t1, Data, views, comment, avtor, like
```

Функция, которая запрашивает данные по странице, возвращает обработанный словарь с данными

```python
def getData(page):
    # запрашиваем данные по ссылке
    response = get(page)

    if not response.ok:
        # если сервер нам отказал, вернем статус ошибки
        return response.status_code
    # получаем содержимое страницы
    html = response.content
    soup = BeautifulSoup(html,'html.parser')

    article = soup.find('h1',{'class':'articleTitle--pUvzk'}).text.replace('\xa0', ' ').strip() #навзание статьи

    t = soup.find('div',{'class':'header--W4Uh8'})
    t1 = t.find('a',{'class':'flow--CBXD2'}).text.replace('\xa0', ' ').strip() #для кого новость

    data = soup.find('div',{'class':'dateWrapper--JWrXQ'})
    Data = data.find('time').text.strip() #дата

    views = soup.find('span',{'class':'counter--6lus3'}).text.strip() #просмотры

    comment = soup.find('span',{'class':'content--FZH2E'}).text.strip() #комментарии
    avtor = soup.find('p',{'class':'paragraph--v7OR4'}).text.strip() #автор
    like = soup.find('span',{'class':'counter--IZ6Up'}).text.strip() #лайки

    r = soup.find_all('p',{'class':'lead--PByZ8'})
    r1 = [item.text.replace('\xa0',' ') for item in r]
    r3 = " ".join(r1).strip() #о чем новость

    r2 = soup.find_all('p',{'class':'paragraph--v7OR4'})
    t = [item.text.replace('\xa0', ' ') for item in r2 ][2:]
    t2 = " ".join(t).strip() #текст новости

    data_row = {"Название":article, "Для кого новость":t1,
                "Дата":Data, "Просмотры":views,
                "Комментарии":comment,
                "Автор":avtor,
                "Лайки":like,
                "О чем новость":r3,"Текст новости":t2}

    return data_row
```

```python
data_row = getData('https://journal.tinkoff.ru/news/binance-leave/')
data_row
```

```python
#соберем  информацию на 40 ссылок
from tqdm import tqdm_notebook

# создаем DataFrame куда кладем всю информацию
final_df = pd.DataFrame(columns=['Название', 'Для кого новость', 'Дата', 'Просмотры', 'Комментарии',
                                 'Лайки', 'О чем новость','Текст новости'])

for page_number in tqdm_notebook(range(40), desc='Pages'):
    #собираем ссылки с текущей страницы
    links = getPageLinks(page_number)

    for link in tqdm_notebook(links, desc='News', leave=False):
    #для каждой ссылки будем это делать 1 раз
        for i in range(1):
            try:
                #пытаемся собрать данные
                data_row = getData(link)
                final_df = final_df.append(data_row, ignore_index=True)
                # если высе получилось - выходим из внутреннего цикла
                break
            except:
                # Иначе, пробуем еще несколько раз, пока не закончатся попытки
                print('AHTUNG! parsing once again:', link)
                continue
```
